```python
import pandas as pd
df = pd.read_csv("/content/50_Startups.csv")
df.head()
```

```python
dummies = pd.get_dummies(df.State)
```

```python
dummies
```

```python
merge=pd.concat([df,dummies],axis='columns')
merge
```

```python
final=merge.drop(['State'],axis='columns')
final.head()
```

```python
#dropping a column
final=final.drop(['New York'],axis='columns')
final.head()
```
