#Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
import numpy as np
from matplotlib import pyplot as plt

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [56, 39, 34, 34, 29]
security = [44 ,36 ,55, 50, 42]

# create an array of len(langauges) size
x_axis = np.arange(len(languages))

#suppose x_axis = [0,1,2,3,4] then for the 1st bar graph first bar graph will draw at -0.2 (0-0.2) and second at 0.8(1-0.2)
#for secomd bar graph first bar will draw at 0.2 2nd at 1.2 , 3rd at 3.2 and so on
plt.bar(x_axis-0.2, popularity , label="Popularity" , color="#800080" ,width=0.4)
plt.bar(x_axis+0.2, security , label="Security" , color="#ff471a", width=0.4)

plt.xticks(x_axis, languages)
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar graph with Multiple bar ")
plt.legend()
plt.show()
```

```python
import numpy as np

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [56, 39, 34, 34, 29]
security = [44 ,36 ,55, 50, 42]

# create an array of len(langauges) size
x_axis = np.arange(len(languages))

#suppose x_axis = [0,1,2,3,4] then for the 1st bar graph first bar graph will draw at -0.2 (0-0.2) and second at 0.8(1-0.2)
#for secomd bar graph first bar will draw at 0.2 2nd at 1.2 , 3rd at 3.2 and so on
plt.barh(x_axis-0.2, popularity , label="Popularity" , color="#660033" ,height=0.4)
plt.barh(x_axis+0.2, security , label="Security" , color="#ff471a", height=0.4)

plt.yticks(x_axis, languages)
plt.ylabel("Languages")
plt.xlabel("Number")
plt.title("Bar graph with Multiple bar")
plt.legend()
plt.show()
```
